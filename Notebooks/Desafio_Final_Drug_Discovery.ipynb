**O que é [Drug Discovery](https://en.wikipedia.org/wiki/Drug_discovery)?**<BR>
É um processo que consiste na descoberta de novos medicamentos. Muito antigamente animais e seres humanos faziam descobertas de forma intuitiva, por exemplo plantas medicinais presentes na natureza. Com o avanço da tecnologia foi possível desenvolver ferramentas e aprimorar o método científico para descobrir e sintetizar novos medicamentos. <BR>
Ainda assim os primeiros medicamentos apresentaram grandes problemas, houve até mortes e grávidas que tiveram o nascimento de seus bebês com deformidades. [*](https://www.ufg.br/n/99956-evolucao-dos-medicamentos-e-retratada-em-palestra)
Antes de chegar no produto final que é o medicamento, existem uma série de passos e o primeiro deles se trata da análise de compostos e sua interação com vários tipos de células, podendo ser as células que o medicamento deseja inibir ou até mesmo células do nosso organismo que não devem ser afetadas senão seria um efeito colateral não desejado.

**Onde entra Ciência de Dados nesse projeto?**<BR>
O processo de Drug Discovery pode ser muito custoso e levar tempo testando até mais de 10 mil compostos durante as fases iniciais de pesquisa.[*](https://www.bndes.gov.br/wps/portal/site/home/conhecimento/noticias/noticia/novos-medicamentos-cgee)
Requer muitos profissionais para analisar composto por composto e analisar a interação com as células, testar diferentes tipos de doses e a reação ao longo do tempo. Isso gera uma enorme quantidade de dados a serem analisados para buscar um resultado e nesse ponto entra da Ciência de Dados e Inteligência Artificial, principalmente Machine Learning.<BR>
Atualmente com o advento da internet temos uma quantidade massiva de dados disponíveis para serem analisados, e com o avanço do hardware temos poder de processamento suficiente para trabalhar esses dados e utilizar IA para aprender com dados sem necessidade de ter regras bem definidas.[*](https://www.kdnuggets.com/2020/11/data-science-history-overview.html)<BR>
Se dependêssmos somente de mão de obra de profissionais para obter resultados, levaria muito tempo para analisar tantos dados assim, então a Ciência de Dados permite processar e inferir sobre uma grande quantidade de dados, sendo assim conseguimos focar esforços em avançar nas descobertas e deixamos o trabalho pesado para o computador, em outras palavras a ciência de dados não elimina a necessidade de pessoas, ainda precisamos de profissionais para avaliar os resultados e ter o conhecimento da área sendo tratada, por outro lado otimizamos o tempo, reduzimos o custo e aumentamos eficiência com resultados mais precisos . [*](https://www.tessella.com/insights/how-is-machine-learning-accelerating-drug-development)

Para ler os dados vamos utilizar a biblioteca [Pandas](https://pandas.pydata.org/) que é capaz de transformar os dados num arquivo compactado em um DataFrame, estrutura similar a de tabelas no qual conseguimos fazer cálculos e analisar mais facilmente os dados.

````python
#importando biblioteca para leitura dos dados
import pandas as pd
#local dos dados
arquivo = 'https://github.com/marcellapbp/imersao-dados-desafio-final/blob/main/Dados/dados_experimentos.zip?raw=true'
#gerando dataframe, indicando que o arquivo está compactado
df = pd.read_csv(arquivo,
                 compression='zip')
````

Outra forma bastante interessante de analisar a correção de variáveis é com um heatmap, pois com esse tipo de gráfico conseguimos olhar várias variáveis ao mesmo tempo num mapa de calor que nos indica quais variáveis possui alta correção.
Lembrando que duas variáveis estarem relacionadas, não significa que uma gera o valor da outra, em outras palavras correção não implica em causalidade. Para entender melhor isso até de uma forma divertida, esse site [WTF Visualizations](http://viz.wtf/) traz exemplos absurdos de correção onde obviamente uma situação não tem nada a ver com a outra. 

````python
#apenas para experimentação utilizei um exemplo de código da própria documentação
import numpy as np

correlacao = df.loc[:,'g-0':'g-50'].corr()

# Cria uma máscara para não duplicar os resultados (triângulo de cima)
mask = np.triu(np.ones_like(correlacao, dtype=bool))

#aumentando o tamanho do gráfico
f, ax = plt.subplots(figsize=(11, 9))

#cria a cor com uma legenda que indica quanto mais próximo de 1 mais relação existe
cmap = sns.diverging_palette(230, 20, as_cmap=True)

#desenha o gráfico
sns.heatmap(corr, mask=mask, cmap=cmap, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
````

Para o gráfico acima existem variações, é comum ver esse tipo de gráfico completo por quadradinhos formando um retângulo, mas estaria repetindo as informações de forma espelhada com uma linha cortando ao meio com a cor vermelha indicando alta correlação. Essa linha vermelha é o exato ponto onde as próprias variáveis se encontram, g-0 com g-0 e g-1 com g-1 e assim por diante. Então para focar no que interessa e tornar o gráfico mais agradável faz sentido eliminar o triângulo de cima.<BR>
Agora olhando para os resultados, observamos alguns dados com uma leve correlação, mas não há nenhum resultado muito destacado. (com vermelho bem forte)
